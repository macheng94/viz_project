## 2. select common tree type

set top 25 trees type, and cluster other species in type 'others'.

```python
tree25 = df.groupby('spc_common')['tree_id'].nunique().sort_values(ascending=False).head(25)
tree25
```

```python
top25tree = ['London planetree', 'honeylocust', 'Callery pear', 'pin oak', 'Norway maple', \
             'littleleaf linden', 'cherry', 'Japanese zelkova', 'ginkgo', 'Sophora', 'red maple', \
             'green ash', 'American linden', 'silver maple', 'sweetgum', 'northern red oak',\
             'silver linden', 'American elm','maple','purple-leaf plum','swamp white oak',\
             'crimson king maple','Chinese elm','Schubert chokecherry','Japanese tree lilac',\
            ]
```

```python
tree25 = df.loc[df['spc_common'].isin(top25tree)]
others = df.loc[~df['spc_common'].isin(top25tree)]
```

```python
trees = pd.concat([tree25,others])
trees
```

## 3. keep only useful columns, keep only alive trees

```python
trees = trees.loc[trees['status'].isin(['Alive'])]
trees = trees[['tree_id','spc_common','latitude','longitude','x_sp','y_sp']]
```
